```python
df['Super_powers'] = df['Super_powers'].str.split(',')
df['combat_str'] = df['combat_str'].str.split()
df['durability_str'] = df['durability_str'].str.split()
df['height_str'] = df['height_str'].str.split()
df['intelligence_str'] = df['intelligence_str'].str.split()
df['level_str'] = df['level_str'].str.split()
df['power_str'] = df['power_str'].str.split()
df['speed_str'] = df['speed_str'].str.split()
df['speed_velocity_str'] = df['speed_velocity_str'].str.split()
df['strength_str'] = df['strength_str'].str.split()
df['strength_force_str'] = df['strength_force_str'].str.split()
df['weight_str'] = df['weight_str'].str.split()
df['tier_str'] = df['tier_str'].str.split()
```

```python
df['tags'] = df['Super_powers'] + df['combat_str'] + df['durability_str'] + df['height_str'] + df['intelligence_str'] + df['level_str'] + df['power_str'] + df['speed_str'] + df['speed_velocity_str']+ df['strength_str'] + df['strength_force_str'] + df['weight_str'] + df['tier_str']
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df['tags'][112]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=1000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity
```

```python
def recommend_heroes(mh):
    mh_index = df[df['name'] == mh].index[0]
    distances = similarity[mh_index]
    hero_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])
    return [x[0] for x in hero_list]
    
```

```python
df['name'].tolist()
```

```python
user_input = 'Batman'
```

```python
custom_index_order = recommend_heroes(user_input)
alignment_mh = df[df['name'] == user_input]['Alignment']
alignment_mh=alignment_mh.iloc[0]
```
