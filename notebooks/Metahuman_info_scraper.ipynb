### Title Extraction Sups

```python
titles = []
for j in range(len(characters)):
    print(j)
    url = f'https://superheroes.fandom.com/wiki/Category:Superheroes?from={characters[j]}'
    response = requests.get(url)
    page_contents = response.text
    doc = BeautifulSoup(page_contents, 'html.parser')
    li_tags = doc.find_all('li', {'class': 'category-page__member'})
    # Getting title names
    for i in range(1,len(li_tags)):
        div_tag = li_tags[i].find('div')
        if div_tag and div_tag.a and div_tag.a.get('title'):
            titles.append(div_tag.a.get('title'))
        elif li_tags[i].a and li_tags[i].a.get('title'):
            titles.append(li_tags[i].a.get('title'))
```

```python
def process_string(s):
    # Remove content inside parentheses
    s = re.sub(r'\([^)]*\)', '', s)
    # Remove roman characters after a word
    s = re.sub(r'\b[IVXLCDM]+\b', '', s)
    # Remove non-alphabetic characters and spaces
    s = re.sub(r'[^a-zA-Z]+', '', s)
    # Convert to lowercase and remove spaces
    s = s.lower()
    return s
```
